```python
# add the ieml library in path to be able to import it
import sys
import os
sys.path.append(os.path.dirname(os.getcwd()))
```

```python
from ieml import IEMLDatabase

# download and load ieml dictionary
ieml_db = IEMLDatabase()
dictionary = ieml_db.dictionary()
```

```python
# display scripts defined in dictionary and theirs frenchs translations
for script in dictionary.scripts:
    print(script, dictionary.translations[script].fr)
```

```python
# relations matrix
R = dictionary.relations.matrix

from ieml.dictionary.script import script

s0 = script("l.a.-t.a.-f.o.-'")
s1 = script("l.a.-d.a.-f.o.-'")
assert all(s in dictionary.scripts for s in (s0, s1))

print(s0, dictionary.translations[s0].fr)
print(s1, dictionary.translations[s1].fr)
print('distance:', R[dictionary.index[s0], dictionary.index[s1]])
```

```python
# sort dictionary with respect to s0
scripts = sorted(dictionary.scripts, key=lambda s1: -R[dictionary.index[s0], dictionary.index[s1]])

#display 50 closest script from s0
for s in scripts[:50]:
    print(R[dictionary.index[s0], dictionary.index[s]], s, dictionary.translations[s].fr)
```
